# **Student Grade Management System**

Create a Python class named Student.
Requirements:
- Create a class attribute named school
- Use the __init__() method to take:
    - Student name
    - Student marks
- Create an instance method named check_grade() that displays:
    - A Grade : Marks >= 80
    - B Grade : Marks >= 60 and < 80
    - C Grade : Marks >= 40 and < 60
    - Fail: Marks < 40
- Create a class method to display the school name
- Create a static method to display a welcome message
- Create an object and display the student's grade

```python
class Student:
    school = "Skill Shikshya"
    # instance method (constructor)
    def __init__(self,name,marks): 
        self.name = name
        self.marks = marks
    
    #instance method 
    def check_grade(self):
        if 100>= self.marks >= 0:
            if 100 >= self.marks >= 80: 
                return "A"
            elif 80 > self.marks >= 60:
                return "B"
            elif 60 > self.marks >= 40:
                return "C"
            else:
                return "Fail"
        else:
            return "Not a valid marks"
    
    # class method
    @classmethod
    def display_name(cls): # here we are passing the student class as cls
        return cls.school

    # static method
    @staticmethod
    def welcome():
        print("Welcome Aspirant")
```

```python
student_name = input("Enter the name of student: ")
student_marks = int(input(f"Enter the marks obtained by {student_name}: "))
student = Student(student_name,student_marks)
student.welcome()
print(f"\nInstitue Name: {student.display_name()}")
print(f"Mr./Mrs.{student_name} has obtained {student.check_grade()} grade.")
```

# **Bank Account Management System**

Create a Python class named BankAccount

Requirements:
- Create a class attribute named bank_name
- Use the __init__() method to take:
    - account holder name
    - initial balance
- Create the following instance methods
    - deposit() => Ask the user for an amount and add it to balance
    - withdraw() => ask the user for an amount and deduct it from the balance if sufficient balance is available
    - show_balance() => Display the current balance
- Create a class method to display the bank name
- Create a static method to display banking rules or a welcome message
- Create an object and test all methods

```python
class BankAccount:
    bank_name = "Nabil Bank"
    # constructor
    def __init__(self,account_holder,initial_balance):
        self.account_holder = account_holder
        self.initial_balance = initial_balance
    
    # instance methods

    # function to deposit
    def deposit(self,deposit_amt):
        self.initial_balance += deposit_amt
        return self.initial_balance
    
    # function to withdraw
    def withdraw(self,withdraw_amt):
        self.initial_balance -= withdraw_amt
        return self.initial_balance
    
    # function to display balance
    def show_balance(self):
        return self.initial_balance

    # class method to print the bank name
    @classmethod
    def print_name(cls):
        return cls.bank_name

    # static method to display bank rule
    @staticmethod
    def welcome_message():
        print("\t\tWelcome to Nabil Bank!!!")
        print("Banking Rules:")
        print("1. Balance must be greater than 0 and sufficient for withdrawal.")
        print("2. Withdrawal amount must be greater than 0 and less than or equal to balance.")
        print("3. Deposit amount must be greater than 0.\n")
```

```python
import time
account_holder = input("Enter the name of account holder: ")
initial_balance = float(input(f"Enter {account_holder}'s current balance: "))
bank_account = BankAccount(account_holder,initial_balance)
while True:
    print(f"\nBank Name: {bank_account.bank_name}\n")
    bank_account.welcome_message()
    
    print("=="*40)
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")
    user_choice = int(input("Enter your choice: "))

    if user_choice == 1:
        deposit_amt = float(input("Enter the amount to deposit: "))
        if deposit_amt > 0: 
            print(f"Successfully deposited Rs.{deposit_amt}.")
        else:
            print(f"Cannot deposit amount less than or equal to 0.")
    elif user_choice == 2:
        withdraw_amt = abs(float(input("Enter the amount to withdraw: ")))
        if withdraw_amt <= bank_account.initial_balance:
            new_balance = bank_account.withdraw(withdraw_amt)
            print(f"Successfully withdrawn Rs.{withdraw_amt}.")
        else:
            print(f"Your balance doesnot meets the withdraw amount.")
    elif user_choice == 3:
        print(f"Your current balance is Rs.{bank_account.initial_balance}.")
    elif user_choice == 4:
        print("Exitting....")
        time.sleep(4)
        break
    else:
        print("Enter a valid choice!!!")
        continue
```
