**Q1. Temperature Converter**
**Create a class Temperature:**
- Constructor accepts celsius
- Create a property fahrenheit
- fahrenheit should automatically convert Celsius to Fahrenheit

```python
class Temperature:
    f_temp = 0
    def __init__(self,c_temp):
        self.c_temp = c_temp
    @property
    def fahrenheit(self):
        self.f_temp = (self.c_temp * 1.8) + 32
        return self.f_temp
```

```python
c_temp = float(input("Enter the temperature(in Celsius): "))
temp = Temperature(c_temp)
print(f"{c_temp} degree in Celsius = {temp.fahrenheit} degree in Fahrenheit")
```

**Q2. Employee Salary Validation**
**Create a class Employee:**
- Constructor accepts salary
- Create a property salary
- Salary cannot be negative
- If a negative salary is assigned, raise a ValueError.

```python
class Employee:
    def __init__(self, sal):
        self._salary = sal

    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError("Salary cannot be negative.")
        self._salary = value
```

```python
sal = float(input("Enter the salary of employee: "))
emp = Employee(sal)
print(f"Salary Amount: Rs.{emp.salary}")
```

**Q3. Bank Account**
**Create a class BankAccount:**
- Constructor accepts balance
- Create a property balance
- Balance cannot be negative
- Create a deleter that prints "Account Closed" when the balance is deleted

```python
class BankAccount:
    def __init__(self, bal):
        if bal < 0:
            raise ValueError("Balance cannot be negative.")
        self._balance = bal

    @property
    def balance(self):
        return self._balance

    @balance.deleter
    def balance(self):
        print("Account closed")
        del self._balance
```

```python
bal = float(input("Enter the balance of bank account holder: "))
bank_account = BankAccount(bal)
print(f"Balance: Rs.{bank_account.balance}")
print(bank_account.bal_deleter)
```
