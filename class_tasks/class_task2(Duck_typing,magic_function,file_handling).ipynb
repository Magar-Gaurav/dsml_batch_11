**1. Write a python program to count the total number of line, words and characters in a text file.**

```python
try:
    file_name = input("Enter the name of file whose line you want to count: ").strip()
    with open(file_name,"r") as f:
        a = f.readline()
        count = 0
        while a:
            count += 1
            a = f.readline()
        print(f"Total number of line on {file_name} is {count}.")
except FileNotFoundError:
    print("File not found!!!")
```


**2. Write a python prorgam to copy the contents of one file into a another file.**

```python
try:
    file_name = input("Enter the name of file whose data you want to copy:").strip()
    content = ""
    with open(file_name,"r") as f:
        content = f.read()
    new_file_name = input(f"Enter the file name in which you want to copy content of {file_name}: ").strip()
    with open(new_file_name,"w") as f:
        f.write(content)
    print(f"Successfully copied content of {file_name} into {new_file_name}.")
except FileNotFoundError:
    print("Error: File not found!!!")
```

**3. Write a python program to replace every occurance of a specific word in a file with another word and save the changes**

```python
try:
    file_name = input("Enter the name of file whose line you want to count: ").strip()
    find_word = input("Enter the word to replace: ").strip().lower()
    replace_word = input("Enter the word which to use for replacement: ").strip().lower()
    with open(file_name,"r") as f:
        a = f.read()
        if find_word in a:
            a = a.replace(find_word,replace_word)
            with open(file_name, "w") as new_f:          
                new_f.write(a)
            print(f"Successfully replaced {find_word} with {replace_word}.")
        else:
            print(f"No {find_word} found!!!")
except FileNotFoundError:
    print("File not found!!!")
```

**4. Create two unrelated classes that both contain a speak() method. Write a function that accepts an object and calls its speak() method. Demonstrate Duck Typing**

```python
class Teacher:
    def speak(self):
        print(f"Mr.Hari is speaking.")
class Student:
    def speak(self):
        print("This student named Ram is speaking.")

def speak_call(object):
    object.speak()
```

```python
teacher = Teacher()
student = Student()
talk_person = input("Who is speaking? (t for teacher, s for student): ").strip().lower()
if talk_person == "t":
    speak_call(teacher)
elif talk_person == "s":
    speak_call(student)
else:
    print("Bro enter the correct options!!!")
```


**5. Create three classes: Emailnotification, SMSNotification, and PushNotification, each having a send() method. Write a common function that can send notifications using any of these objects.**

```python
class EmailNotification:
    def send(self):
        print("Brrrr!!!\nSent notification through email.")
class SMSNotification:
    def send(self):
        print("Brr!!!\nSent notification through SMS.")
class PushNotification:
    def send(self):
        print("Brr!!!\nSent notification through Push Notification.")
# duck typing function
def send_notification(object_name):
    object_name.send()
```

```python
email = EmailNotification()
sms = SMSNotification()
push = PushNotification()
print("1. For email (e)")
print("2. For sms (s)")
print("3. For push (p)")
user_choice = input("Enter your choice: ").strip().lower()
if user_choice == "e":
    send_notification(email)
elif user_choice == "s":
    send_notification(sms)
elif user_choice == "p":
    send_notification(push)
else:
    print("Bruh!!! Enter a valid option.")
```

**6. Create a Student class with marks in two subjects. Implement the __add__() magic method so that adding two student objects returns a new student object with combined marks.**

```python
class Student:
    def __init__(self,mark1,mark2):
        self.mark1 = mark1
        self.mark2 = mark2

    def __add__(self,other):
        self.mark1 = self.mark1 + other.mark1
        self.mark2 = self.mark2 + other.mark2
        return self.mark1,self.mark2
 
    def __repr__(self):
        return f"Student with combined marks: {self.mark1,self.mark2}"
```

```python
student1 = Student(10,20)
student2 = Student(30,40)
print(f"Student with combined marks of student1 and student2 is {student1 + student2}.")
```

**7. Create a point class with x and y coordinates. Implement:**
- __str__() to display the point in a readable format
- __add__() to add two points together
- __repr__() to return a developer-friendly representation of the object

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __str__(self):
        return f"Value of x = {self.x}\nValue of y = {self.y}"
    def __add__(self,other):
        return (self.x + other.x,self.y + other.y)
    def __repr__(self):
        return f"Value of x coordinate = {self.x}\nValue of y coordinate = {self.y}"
```

```python
point1 = Point(2,3)
point2 = Point(4,5)
print("Printing details using str:")
print(f"Details of point1:\n{str(point1)}")
print(f"Details of point2:\n{str(point2)}")
print("\nPrinting details using repr: ")
print(f"Sum of point1 and point2:\n{point1 + point2}")
print(f"Details of point1:\n{repr(point1)}")
print(f"Details of point2:\n{repr(point2)}")
```
