```python
# creating dictionary using curly brackets
dict_1 = {
    'id' : 101,
    'name' : 'Saurav',
    'email' : 'sauravmagar664@gmail.com',
    'role' : 'Data Scientist'
}
print(type(dict_1))
print(dict_1)
```

```python
# creating dictionary using dict constructor
new_dict = dict({
    'name' : 'Gaurav',
    'cast' : 'Magar',
    'profession' : 'Student'
})
print(type(new_dict))
print(new_dict)
```

```python
# dictionary inside a dictionary
new_dict = dict({'student1' : {'name' : 'Dhiraj','age' : '22'},'student2' : {'name': 'Ram', 'age' : 21}})
print(new_dict)
```

```python
# creating dictionary from sequence having each item as a pair
person = dict([('name','rita'),('country','nepal'),('telephone','9898000000')])
print(person)
```

```python
# accessing values
data_new = {
    'name' : 'Saurav',
    'age' : 23
}
print(data_new['name'])
print(data_new['age'])
```

```python
# accessing using .get()
data_new.get('name')
```

```python
# manipulating values (modifying)
data_new['name'] = 'Gaurav'
data_new
```

```python
# using keys, values and items to retrive values
print(data_new.keys())
print(data_new.values())
print(data_new.items())
```

```python
# pop => returns and removes the item with key and return its value
print(data_new.pop('name'))
```

```python
# popitem => returns and remove the last indexed item from the dictionary
print(data_new.popitem())
```

```python
# del key => delete the item with the key that is passed
del data_new['name']
print(data_new)
```

```python
# clear() => removes all the items from the dictionary. Empty the dictionary
data_new.clear()
print(data_new)
```

```python
# del dict_name => deletes the entire dictionary
del data_new
```

```python
# update() => used to join two dictionary
dict1 = {'name' : 'Saurav'}
dict2 = {'age' : 23}
dict1.update(dict2)
print(dict1)
```

```python
# **kwargs => used to pass multiple parameters (in dictionary we can used to combine multiple dictionary)
dict3 = {'id' : 10001}
dict4 = {'email' : 'sauravmagar664@gmail.com'}
dict4 = {**dict1 , **dict2, **dict3}
print(dict4)
```

```python
# sorted => used to sort keys of a dictionary in ascending order by default
student = {'b' : 2,'a' : 1,'c' : 3}
result = sorted(student)
print(result)
```

```python
# Create a dictionary to store student name, age and faculty
student_data = {
    'name' : 'Saurav',
    'age' : 23,
    'faculty' : 'BCA',
    'city' : 'Kathmandu'
}
print(student_data)
```

```python
# Add a new key called phone to an existing dictionary
student_data['phone'] = 9744445970
print(student_data)
```

```python
# Update the value of age in a dictionary
student_data['age'] = 24
print(student_data)
```

```python
# Remove a key named city from a dictionary using del
del student_data['city']
print(student_data)
```

```python
# check whether the key email exists in a dictionary
if 'email' in student_data:
  print(f"Yes, {student_data['name']}'s email exists.")
else:
  print(f"No, {student_data['name']}'s email doesnot exists.")
```
