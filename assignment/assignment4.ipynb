**Project Title:**

Student Management System using OOP Concepts

**Problem Statement:**

Develop a Student using Python that demonstrates the concepts of Class, Object, Encapsulation, Inheritance and Abstraction. The system should allow users to store student information, calculate grades, and display student details based on different student categories.

**Requirements:**

**1. Create an Abstract class Student**
The astract class should contain:
- Student ID
- Student Name
- A private attribute __marks
- A method to set marks
- A method to get marks
- An abstract method calculate_grade()
- An method display_details()

```python
from abc import ABC, abstractmethod

class Student(ABC):
  def __init__(self,s_id,s_name,s_marks):
    self.s_id = s_id
    self.s_name = s_name
    self.__s_marks = s_marks
    self.grade = " "

  def get_marks(self):
    return self.__s_marks

  def set_marks(self,marks):
    if 100 >= marks >= 0:
      self.__s_marks = marks
    else:
      print("Invalid Marks. Marks should be between 0 and 100.")

  @abstractmethod
  def calculate_grade(self):
    pass

  def display_details(self):
    print(f"\nStudent Id: {self.s_id}\nStudent Name: {self.s_name}\nStudent Marks: {self.__s_marks}\nGrade Obtained : {self.grade}\n")
```

```python
class DSMLStudent(Student):
  def __init__(self,s_id,s_name,s_marks):
    super().__init__(s_id,s_name,s_marks)
  def calculate_grade(self):
    marks = self.get_marks()
    if marks >= 85:
        self.grade = "A"
    elif marks >= 65:
        self.grade = "B"
    elif marks >= 45:
        self.grade = "C"
    elif marks >= 32:
        self.grade = "D"
    else:
        self.grade = "Fail"
    return self.grade
```

```python
class MERNStudent(Student):
  def __init__(self,s_id,s_name,s_marks):
    super().__init__(s_id,s_name,s_marks)
  def calculate_grade(self):
    marks = self.get_marks()
    if marks >= 85:
        self.grade = "A"
    elif marks >= 65:
        self.grade = "B"
    elif marks >= 45:
        self.grade = "C"
    elif marks >= 32:
        self.grade = "D"
    else:
        self.grade = "Fail"
    return self.grade
```

```python
name = input("Enter the name of student who is studying DSML in SkillShikshya: ")
id = input(f"Enter the id of {name}: ")
marks = int(input(f"Enter the marks obtained by {name}: "))
student1 = DSMLStudent(id,name,marks)
student1.calculate_grade()
student1.display_details()
```
