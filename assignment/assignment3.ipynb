A school has conducted its final examination and scored student names and marks in two seperate lists.

`students = ["Ram","Sita","Hari","Gita","Shyam","Rita"]`

`marks = [85,42,78,91,35,67]`

The school administartion wants a small system that can process and analyze the results automatically. As a python developer, your task is to develop a program that solves the following requirements using Python's built-in functions such as map(), filter(), reduce(), zip() and enumerate().

`Questions`
- Display all student names with their serial numbers starting from 1.
- Combine student names and marks into a single collection containing student records.
- Generate grades for each student (like in SEE)
- Create a list containing only the students who have passed the examination. (marks above 40 is pass)
- Find the highest mark among all students without using the built-in max() function.

```python
students = ["Ram","Sita","Hari","Gita","Shyam","Rita"]
marks = [85,42,78,91,35,67]
```

`Printing name of students with their serial number starting from 1`

```python
for index, name in enumerate(students,start = 1):
  print(f"{index}. {name}")
```

`Combining student names and marks into a single collection containing student records using zip`

```python
single_collection = dict(zip(students,marks))
print(single_collection)
```

`Generating grades for each student like on SEE`

```python
def get_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B+"
    elif marks >= 60:
        return "B"
    elif marks >= 50:
        return "C+"
    elif marks >= 40:
        return "C"
    elif marks >= 30:
        return "D"
    else:
        return "E"
for key, value in single_collection.items():
  grade = get_grade(value)
  single_collection[key] = [value, grade]
# printing all of the grades
for key, value in single_collection.items():
  print(f"Name : {key}  Marks Obtained : {value[0]}  Grade : {value[1]}")
```

`Creating a list of students who have passed the examination (marks above 40 is pass)`

```python
def check_pass(marks):
  name, mark = marks
  return mark[0] > 40
passed_students = dict(filter(check_pass,single_collection.items()))
for key, value in passed_students.items():
  print(f"Name : {key}  Marks Obtained : {value[0]}  Grade : {value[1]}")
```

`Calculate total marks obtained by all students combined`

```python
from functools import reduce
def total_marks(total,datas):
  mark,grade = datas
  return total + mark
total_mark = reduce(total_marks,single_collection.values(),0)
print(f"Total marks obtained by all students : {total_mark}")
```

`Find the highest mark among all student without using the built-in max() function`

```python
def highest_mark(highest,data):
  mark, grade = data
  return mark if mark > highest else highest
mark = reduce(highest_mark,single_collection.values(),0)
print(f"The highest marks obtained was : {mark}")
```
