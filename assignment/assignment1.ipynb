Question number 1: Create a Python program using at least 10 string methods from the table below and show their outputs with examples. Try methods like upper(), split(), replace(), find() and strip()

```python
my_str = "Hello World 123"
```

```python
# 1. upper() => this converts string into the uppercase
print(f"After using upper() method:\n{my_str.upper()}")
```

```python
# 2. lower() => this converts string into the lowercase
print(f"After using lower() method:\n{my_str.lower()}")
```

```python
# 3. capitalize() => this capitalizes the first letter
print(f"After using capitalize() method:\n{my_str.capitalize()}")
```

```python
# 4. title() => this capitalizes the first letter of every word
print(f"After using title() method:\n{my_str.title()}")
```

```python
# 5. strip() => removes whitespaces from both end
my_newstr = " Hello World   "
print(f"Without using strip() method:\n{my_newstr}")
print(f"After using strip() method:\n{my_newstr.strip()}")
```

```python
# 6. replace() => replaces parts of a string
print(f"After replacing 'o' with 'O' using replace method:\n{my_str.replace('o','O')}")
```

```python
# 7. find() => returns the index of first occurance (-1 if not found)
print("Note: If there is no occurance the output will be -1.")
print(f"The index of 'o' using find() method is: {my_str.find('o')}.")
print(f"The index of 'a' using find() method is: {my_str.find('a')}.")
```

```python
# 8. split() => splits string into a list using a string operator
print(f"After splitting on ' ' using split() method:\n{my_str.split(" ")}")
```

```python
# 9. startswith() => checks if string starts with a substring
print(f"Do my_str starts with h?:\t{my_str.lower().startswith('h')}")
```

```python
# 10. endswith() => checks if string ends with a substring
print(f"Do my_str ends with d?:\t{my_str.lower().endswith('d')}")
```

```python
# 11. isalnum() => returns True if all characters are alphanumeric
print(f"Is my_str alphanumeric?:\t{my_str.isalnum()}")
```

```python
# 12. isalpha() => returns True if all characters are alphabetic
print(f"Is my_str is alphabetic?:\t{my_str.isalpha()}")
```

```python
# 13. isdigit() => returns True if all charcters are digits
print(f"Is my_str numeric?:\t{my_str.isdigit()}")
```

```python
# 14. isspace() => returns True if all characters are whitespaces
print(f"Is my_str all spaces?:\t{my_str.isspace()}")
```

```python
# 15. islower() => returns True if all cased characters are lowercase
print(f"Is my_str lowercased?:\t{my_str.islower()}")
```

```python
# 16. isupper() => returns True if all cased characters are uppercase
print(f"Is my_str uppercased?:\t{my_str.isupper()}")
```

Question number 2: 
Take a sentence from user and perform different string operations on it such as converting case, counting words, checking digits/alphabet and replacing words. Print all results clearly with labels.

```python
user_str = input("Enter a string: ")
print(f"User string: \t {user_str}")
print(f"User string in uppercase: \t {user_str.upper()}")
print(f"User string in lowercase: \t {user_str.lower()}")
print(f"Total number of words in User's string: \t {len(user_str)}")
print(f"Is User's string only numeric?: \t {user_str.isnumeric()}")
print(f"Is User's string only alphabetic?: \t {user_str.isalpha()}")
print(f"User's string after replacing all 'a' by '@': \t {user_str.replace('a','@')}")
```
