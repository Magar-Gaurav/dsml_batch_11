# Student Management System

Create a user-defined function called student_result() that:
- Takes the names and marks of 5 students as input
- Stores student data in dictionary
- Uses conditional statements to assign grades:
    - A : 80 and above
    - B : 60 - 79
    - C : 40 - 59
    - F : Below 40
- Displays each student's name, marks and grades
- Calculate and display
    - Highest marks
    - Lowest marks
    - Average marks

```python
def SRMS(student_data):
    print("\nStudent's Data:\n")
    for key, value in student_data.items():
        print(f"Name : {key}\tMarks : {value['marks']}\tGrade : {value['grade']}\n")
    marks_list = []
    for info in student_data.values():
        marks_list.append(info['marks'])
    print(f"Highest marks: {max(marks_list)}")
    print(f"Lowest marks: {min(marks_list)}")
    print(f"Average marks: {sum(marks_list) / len(marks_list)}")

student_data = {}
for i in range(5):
    name = input("Enter the user name: ")
    marks = int(input("Enter the marks obtained: "))
    if marks >= 80:
        grade = "A"
    elif 80 > marks >= 60:
        grade = "B"
    elif 59 > marks > 40:
        grade = "C"
    elif 40 > marks > 0:
        grade = "F"
    else:
        print("Invalid marks")
        break
    student_data[name] = {"marks" : marks, "grade" : grade}
SRMS(student_data)
```

## Smart Shopping Cart System

Create a user-defined function called shopping_cart() that:
- allows the user to enter 5 product prices
- stores the price in a list
- uses a loop to calculate the total bill
- applies discounts using conditional statements:
    - total >= 5000 = 15% discount
    - total >= 3000 = 10% discount
    - total >= 1000 = 5% discount
    - otherwise = no discount
- displays:
    - product lsit
    - total amount
    - discount amount
    - final payable amount

```python
def shopping_cart(product_price,total_price,discount):
    print(f"Your product prices list is:\t{product_price}")
    print(f"Your total_amount is: Rs.{total_price:.2f}.")
    print(f"Discount amount: Rs.{total_price * discount / 100:.2f}.")
    print(f"Final payable amount: Rs.{total_price - (total_price * discount / 100):.2f}.")

product_price = []
total_price = 0
for i in range(5):
    price = float(input(f"Enter the {i + 1}th price: "))
    product_price.append(price)
for i in product_price:
    total_price += i
if total_price >= 5000:
    discount = 15
elif total_price >= 3000:
    discount = 10
elif total_price >= 1000:
    discount = 5
else:
    discount = 0
shopping_cart(product_price,total_price,discount)
```
