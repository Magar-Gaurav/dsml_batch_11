```python
# Write a function that takes three numbers as arguments and return the largest number
def greatest(a,b,c):
  return max(a,b,c)
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))
largest = greatest(num1,num2,num3)
print(f"The largest number among {num1},{num2},{num3} is {largest}.")
```

```python
# Write a function that takes a string and return the numbers of vowel present in it
def count_vowel(user_str):
  vowels = ["a","e","i","o","u"]
  vowel_counter = 0
  for i in user_str.lower():
    if i in vowels:
      vowel_counter += 1
  return vowel_counter
user_str = input("Enter your string: ")
count = count_vowel(user_str)
print(f"The total number of vowel present in '{user_str}' is {count}.")
```

```python
# take 5 user input data in list and then write it on user_data.txt
def write_data():
  user_nums = []
  for i in range(5):
    num = int(input("Enter the number: "))
    user_nums.append(num)
  file = open("user_data.txt",mode = "w")
  file.write(str(user_nums))
  print("Successfully writed your input numbers in the file.")
  file.close()
write_data()
```

```python
# reading the file generated
with open("user_data.txt",mode = "r") as f:
  print(f.read())
```

```python
# a list has number from 1 to 7 need index, values without using range
values = [1,2,3,4,5,6,7]
for index,value in enumerate(values):
  print(index, value)
```

```python
user_dets = {
    "name" : "Saurav",
    "gender" : "male",
    "dob" : "1997-09-13",
    "salary" : 100000
}
for index,value in enumerate(user_dets.items(),start=1):
    print(f"{index}. {value[0]} : {value[1]}")
```

```python
# or
for index, (key,value) in enumerate(user_dets.items()):
  print(f"{key} : {value}")
```

```python
lst1 = ["name","gender","dob","salary"]
lst2 = ["Saurav","male","1997-09-13",100000]
# zipping
lst3 = list(zip(lst1,lst2))
print(lst3)
```
