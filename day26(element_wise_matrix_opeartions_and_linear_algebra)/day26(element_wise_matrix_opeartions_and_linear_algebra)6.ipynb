```python
import numpy as np
```

**Element wise matrix addition**

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
addition = a + b
print("Sum of a and b:\n",addition)
```

**Element wise matrix subtraction**

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
difference = a - b
print("Difference of a and b:\n",difference)
```

**Element wise matrix multiplication**

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
product = np.dot(a,b)
print("Product of a and b matrix:\n",product)
```

```python
new_prod =a @ b
print(new_prod)
```

```python
new_prod = a * b
print(new_prod)
```

```python
mat1 = np.array([[1,2,3]])
mat2 = np.array([[1],[2],[3]])
product = np.dot(mat1,mat2)
normal_product = mat1 * mat2
print("Dot product: ",product)
print("Normal product:\n",normal_product)
```

**Multiplying using different row and column**

```python
mat1 = np.array([1,2,3])
mat2 = np.array([[1,2,3]])
product = mat1 @ mat2
print("Product: \n",product) # this one will result error because the n of mat1 doesnot matches the m of mat2   
```

**Element wise division in matrix**

```python
mat1 = np.array([1,2,3])
mat2 = np.array([[1,2,3]])
division = mat1 / mat2
print("Division of mat1 and mat2:\n",division)
```

```python
mat1 = np.array([[1,2,3]])
mat2 = np.array([[1],[2],[3]])
division = mat1 / mat2
print("Division of mat1 and mat2:\n",division)
```

# **Scalar**

```python
scalar = 2
mat1 = np.array([[1,2,3]])
result_add = mat1 + scalar
print("After adding by scalar value:\n",result_add)
result_product = mat1 * scalar
print("After multiplying by scalar value:\n",result_product)
result_difference = mat1 / scalar
print("After dividing by scalar value:\n",result_difference)
```

# **Linear Algebra**

linear algebra means mathematical expression.Eg:
- 2x + y = 5
- x + 3y = 6

```python
A = np.array([[2,1],
              [1,3]
            ])
B = np.array([5,6])
x = np.linalg.solve(A,B)
print(f"x: {x[0]}\ny: {x[1]}")
```

```python
A = np.array([[1,1,2,1,1],[2,-1,1,3,2],[1,2,1,-1,1],[3,1,2,1,-1],[1,-1,1,2,3]])
B = np.array([5,8,4,10,7])
X = np.linalg.solve(A,B)
print("Value of u:",X[0])
print("Value of v:",X[1])
print("Value of x:",X[2])
print("Value of y:",X[3])
print("Value of z:",X[4])
```

```python
x = np.array([[1,1],[2,1],[3,1]])
y = np.array([[2],[3],[5]])
final_product = np.linalg.inv(x.T @ x) @ x.T @ y
print("Slope (m):",final_product[0])
print("Interception (c):",final_product[1])
```

**Determinant of a matrix**

```python
a = np.array([[1,2],[4,5]])
print("Determinant of a: ",np.linalg.det(a))
print("Inverse of a:\n",np.linalg.inv(a))
```

**Eigen Values and Eigen Vectors**

```python
a = np.array([[1,2],[4,5]])
eigen = np.linalg.eig(a)
print("Eigen values:",eigen[0])
print("Eigen vector:\n",eigen[1])
```

**SVD (Single Value Decomposition)**

```python
u,s,v = np.linalg.svd(a)
print("Value of u:\n",u)
print("Value of s:\n",s)
print("Value of v:\n",v)
```
