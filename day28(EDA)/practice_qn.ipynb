# **Ticket Fare Analysis**

**The shipping company wants to understand ticket pricing.**

**Tasks:**

- Find average fare paid by the passengers
- Find the highest and lowest fare
- Determine which passenger paid the highest fare

average fare paid by passengers

```python
df.groupby("pclass")["fare"].mean(numeric_only = True)
```

**highest and lowest fare**

```python
highest_fare = df.groupby("pclass")["fare"].max()
lowest_fare = df.groupby("pclass")["fare"].min()

print("Highest Fare:\n",highest_fare)
print("Lowest Fare:\n",lowest_fare)
```

**highest fare passenger**

```python
df.loc[df["fare"] == df["fare"].max()]
```
