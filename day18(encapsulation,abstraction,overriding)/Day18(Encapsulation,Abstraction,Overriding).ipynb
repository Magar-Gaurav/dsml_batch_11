## **Class Task**

**Create a class BankAccount that:**
- Allows a user to create an account by providing a name and an initial balance. Let balance be private variable.
- Includes methods to deposit, withdraw and check balance. The withdrawl should not allow an account to go negative
- Uses inheritance to create a subclass SavingsAccount that adds an interest rate. Implement a method to calculate interest on the balance and update it accordingly

```python
class BankAccount:
  def __init__(self,name,balance = 0,amt = 0,w_amt = 0):
    self.name = name
    self.initial_balance = balance
    self.amount = amt
    self.w_amt = w_amt
  def deposit(self):
    if self.amount > 0:
      self.initial_balance += self.amount
      print(f"Successfully deposited Rs.{self.amount}.")
    else:
      print("Cannot deposit amount less that equal to 0.")

  def withdraw(self):
    if self.initial_balance >= self.w_amt or self.initial_balance > 0:
      self.initial_balance -= self.w_amt
      print(f"Successfully withdrawn Rs.{self.w_amt}.")
    else:
      print("Your balance doesnot matches your withdraw amount.")

  def check_balance(self):
    print(f"Your balance is {self.initial_balance}.")

class SavingsAccount(BankAccount):
  def __init__(self,name,balance,amt,w_amt,rate):
    super().__init__(name = name,balance = balance,amt = amt,w_amt = w_amt)
    self.rate = rate
  def calc_interest(self):
    interest = self.initial_balance * self.rate
    print(f"The interest amount of your balance is {interest}.")
```

```python
name = input("Enter your name: ").strip().title()
balance = float(input("Enter your initial balance: "))
amt = 1000
w_amt = 2000
rate = 3
s1 = SavingsAccount(name,balance,amt,w_amt,rate)
s1.deposit()
s1.withdraw()
s1.check_balance()
s1.calc_interest()
```

# **Method Overriding**

```python
class Animal:
  def make_sound(self):
    print("Animal is making sound")

class Dog(Animal):
  def make_sound(self):
    super().make_sound()
    print("Dog is barking")
```

```python
dog = Dog()
dog.make_sound()
```

# Manually initializing parent attributes

```python
class Employee:
  def __init__(self,nm,sal):
    self.name = nm
    self.sal = sal
  def getName(self):
    return self.name
  def getSalary(self):
    return self.sal

class SalesOfficer(Employee):
  def __init__(self,nm,sal,inc):
    super().__init__(nm,sal)
    self.inc = inc
  def getSalary(self):
    return self.salary + self.inc
```

```python
e1 = Employee("Rajesh",9000)
print('Total salary of {} is Rs. {}'.format(e1.getName(), e1.getSalary))
```
