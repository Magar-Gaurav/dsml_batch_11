# **Method Overloading**

**Example of creating calculator with two method with same name but different parameters.**

```python
class Calculator:
  def add(self,a,b):
    return a + b
  def add(self,a,b,c):
    return a + b + c
```

```python
calc = Calculator()
calc.add(1,2,3)
```

**Real Example of Method Overloading**

- *Using default value*

```python
class Calculator:
  def add(self,a,b,c=0):
    return a + b + c
```

```python
obj1 = Calculator()
obj1.add(1,2,3)
```

```python
obj2 = Calculator()
obj2.add(1,2)
```

- *Using args*


```python
class Calculator:
  def add(self,args):
    return sum(args)
```

```python
n = int(input("Enter the number of items you will add: "))
nums_list = []
for i in range(n):
  num = int(input("Enter the number: "))
  nums_list.append(num)
calc = Calculator()
print(calc.add(nums_list))
```

# **Inheritance**

*Single Inheritance*

```python
class Animal:
  def sound(self):
    print("Animal makes sound")

class Dog(Animal):
  def bark(self):
    print("Dog is barking.")
```

```python
dog = Dog()
dog.sound()
dog.bark()
```

```python
a = Animal()
a.sound()
```

*Multiple Inheritance*

```python
class Father():
  def work(self):
    print("Father works in a company.")

class Mother():
  def profession(self):
    print("Mother works as a housewife.")

class Child(Father, Mother):
  def occupation(self):
    print("Child is currently reading.")
```

```python
child = Child()
child.occupation()
child.work()
child.profession()
```

*Multi-Level Inheritance*

```python
class GrandParent:
  def grand_intro(self):
    print("I am grandfather.")
class Parent(GrandParent):
  def parent_intro(self):
    print("I am parent.")
class Child(Parent):
  def child_intro(self):
    print("I am child.")
```

```python
child = Child()
child.child_intro()
child.parent_intro()
child.grand_intro()
```

*Hierarchial Inheritance*

```python
class Parent:
  def parent_info(self):
    print("Hey I am parent.")
class Child1(Parent):
  def child1_info(self):
    print("Hey I am child1.")
class Child2(Parent):
  def child2_info(self):
    print("Hey I am child2.")
```

```python
child1 = Child1()
child2 = Child2()
child1.child1_info()
child1.parent_info()
child2.child2_info()
child2.parent_info()
```

*Hybrid Inheritance*

```python
class GrandParent:
  def grand_intro(self):
    print("Hey I am grandparent.")
class Father(GrandParent):
  def father_info(self):
    print("Hey I am father")
class Mother:
  def mother_info(self):
    print("Hey I am mother")
class Child(Father,Mother):
  def child_info(self):
    print("Hey I am child")
```

```python
child = Child()
child.child_info()
child.mother_info()
child.father_info()
child.grand_intro()
```

Implement methods in each class and show that a TeachingAssistant object can access features from both Student and Teacher

```python
class Person:
  def person_info(self,name,age):
    print(f"{name} is {age} years old.")

class Student(Person):
  def student_info(self,student_id):
    print(f"Student Id: {student_id}")

class Teacher(Person):
  def teacher_info(self,subject):
    print(f"Subject: {subject}")

class TeachingAssistant(Student,Teacher):
  def ta_info(self,department,salary):
    print(f"This teaching assistant is for {department} Department. This person's salary is Rs.{salary}.")
```

```python
ta = TeachingAssistant()
ta.ta_info("IT",40000)
ta.teacher_info("Computer Science")
ta.student_info("1001")
ta.person_info("Saurav",23)
```

*Constructor (Super Class)*

```python
class OTPsubscription:
  def __init__(self,subscription_id,plan,total_payment):
    self.subscription_id = subscription_id
    self.plan = plan
    self.total_payment = total_payment

  def subscription(self):
    print(f"Subscriber with {self.subscription_id} subscribed to {self.plan}.")

  def unsubscribe(self):
    print(f"Subscriber with {self.subscription_id} unsubscribed to the {self.plan}.")
```

```python
class PremiumSubscription(OTPsubscription):
  def __init__(self,subscription_id,plan,total_payment,screens):
    super().__init__(subscription_id,plan,total_payment)
    self.max_screens = screens
  def set_max_screen(self,screens):
    self.max_screens = screens
    print(f"Maximum screen set to {self.max_screens} in the premium plan..")
```

```python
obj = PremiumSubscription(1001,"Pro","10000",5)
obj.subscription()
obj.set_max_screen(4)
obj.unsubscribe()
```

# *Same Class Task*

```python
class Person:
  def __init__(self,name,age,**kwargs):
    self.name = name
    self.age = age
    super().__init__(**kwargs)
  def person_details(self):
    print(f"{self.name} is {self.age} years old.")

class Student(Person):
  def __init__(self,name,age,student_id,**kwargs):
    super().__init__(name,age,**kwargs)
    self.student_id = student_id
  def student_details(self):
    print(f"Student id: {self.student_id}\nName: {self.name}\nAge: {self.age}")

class Teacher(Person):
  def __init__(self,name,age,subject,**kwargs):
    super().__init__(name,age,**kwargs)
    self.subject = subject
  def teacher_details(self):
    print(f"{self.name} teaches {self.subject}.")

class TeachingAssistant(Student,Teacher):
  def __init__(self,name,age,student_id,subject,department,salary):
    super().__init__(name=name,age=age,student_id=student_id,subject=subject)
    self.salary = salary
    self.department = department
  def ta_details(self):
    print(f"{self.name} is from {self.department} and has salary Rs.{self.salary}.")
```

```python
student = Student("Saurav",23,101)
student.student_details()
```

```python
teacher = Teacher("Ram",32,"DSA")
teacher.teacher_details()
```

```python
ta = TeachingAssistant("Sita",30,101,"DSA","CSE",1000000)
ta.ta_details()
```
