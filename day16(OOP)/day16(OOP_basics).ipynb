# Example of class and object

```python
# class name is fruits
class Fruits:
  name = "mango" # name is an attribute which stores value (mango)
```

```python
data = Fruits() # object of Fruits class is created
data.name # accessing name attribute using object
```

```python
data2 = Fruits()
data2.name = "Apple" # changing the value of object
data2.name # prints Apple because it overrides class attribute
```

# Class attributes

```python
class Student:
  '''Student Information'''
  school = "Skill Shikshya" # class attribute
  @classmethod # decorator
  def get_school(cls): # instead of cls we can use anywords but we dont use because it is convention not compulsory
    print(f"Currently studying in {cls.school}.")
std = Student()
std.get_school()
```

# Instance method

```python
class Calculator:
  def __init__(self,a,b):
    self.a = a
    self.b = b
  def add(self):
    return self.a + self.b
calculator = Calculator(5,10)
calculator.add()
```

# Class method

```python
class Calculator:
  a = 5
  b = 10
  @classmethod
  def add(cls):
    return cls.a + cls.b
Calculator.add()
```

# Static method

```python
class Dog:
  count = 0
  def __init__(self, name, age = 3):
    self.name = name
    self.age = age
    Dog.count += 1
  def bark(self):
    print(f"{self.name} is barking.")
  @classmethod
  def get_count(cls):
    return cls.count
  @staticmethod
  def get_vet_info():
    return "Dr. Ram K.C"
```

```python
dog = Dog("Tommy",10)
print(f"Name of dog: {dog.name}")
print(f"Age of dog: {dog.age}")
dog.bark()
print(f"Number of dog in vetnary is {dog.get_count()}.")
print(f"{dog.get_vet_info()} is checking up {dog.name}.")
```

# Class Task: Calculator using different types of attributes and methods
- Create a python class named Calculator that demonstrates the use of different types of attributes and methods:
- Requirements:
  - Create a class attribute named company and assign it any company name
  - Use the __init__() method to:
    - Ask the user to enter two numbers
    - Ask the user to enter an operator(+,-,*,/)
  - Create an instance method named calculate() that performs the appropriate opeartion based on the operator entered by the user
  - Create a class method named show_company() that displays the company name
  - Create a static method named welcome() that displays a welcome message
  - Create an object of the class and call all the required methods

```python
class Calculator:
  company_name = "Casio" # class attribute
  def __init__(self,num1,num2,operator): # instance method
    self.num1 = num1
    self.num2 = num2
    self.operator = operator
  def calculate(self): # instance method
    if self.operator == "+":
      return self.num1 + self.num2
    elif self.operator =="-":
      return self.num1 - self.num2
    elif self.operator == "*":
      return self.num1 * self.num2
    elif self.operator == "/":
      if self.num2 != 0:
        return self.num1 / self.num2
      else:
        return "Cannot divide by Zero."
    elif self.operator == "%":
      if self.num2 != 0:
        return self.num1 % self.num2
      else:
        return "Cannot divide by Zero."
    else:
      return "Please enter a valid opeartor."
  @classmethod
  def show_company(cls): # class method
    return cls.company_name
  @staticmethod # static method
  def welcome():
    return "Welcome user. Hope you will enojy it."
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter the opeartor you want to perform (+, -, *, /,%): ")
calc = Calculator(num1,num2,operator)
calc.welcome() # calling static method (it welcomes user)
calc.show_company() # calling class method (it displays company name)
print(f"{num1} {operator} {num2} = {calc.calculate()}") # calling instance method (it will print the output of the operation to be performed)
```
