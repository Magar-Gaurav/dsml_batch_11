```python
# write a python program using user definde function:
# takes two numbers from the user
# takes an opertor sign
# performs selected operation
# displays the result
def Calculator(a,b,operator):
  if operator == "+":
    return a + b
  elif operator == "-":
    return a - b
  elif operator == "*":
    return a * b
  elif operator == "/":
    return a / b
  else:
    print("Please input a valid opeator ( + , -, * , /)")
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
operator = input("Enter operator to perform operation(+,-,*,/):")
calculation = Calculator(a,b,operator)
print("Output:\n",calculation)
```

```python
# Write a function area(length, breadth) that returns the area of rectangle
def Area(length,breadth):
  return length * breadth
length = float(input("Enter the value of length: "))
breadth = float(input("Enter the value of breadth: "))
print(f"The area is {Area(length,breadth)} meter square.")
```

```python
# Write a user-defined function that takes 5 user inputs:
# if user enters an integer add it to a list
# if user enters a float number, remove the last element
# Finally display the updated list after each operation
def user_input():
  user_nums_list = []
  for i in range(5):
    num = input("Enter a number: ")
    if num.isdigit():
      user_nums_list.append(int(num))
    elif "." in num:
      user_nums_list.pop()
    print(user_nums_list)
user_input()
```
