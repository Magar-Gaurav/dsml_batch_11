# **Pandas Visualization**

**importing pands and matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**mouting the drive to use the file stored in google drive**

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
titanic_df = pd.read_csv("/content/drive/MyDrive/dsml/titanic.csv")
```

```python
print("Column names:\n",titanic_df.columns)
```

**Bar chart**

```python
titanic_df["sex"].value_counts().plot.bar()
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

**Survival count**

```python
titanic_df["survived"].value_counts().plot(
    kind = "bar",
    title = "Count of Survived and Dead passengers",
    figsize = (10,6),
    legend = True,
    yticks = list(range(0,600,50)),
    xlabel = "Survived (0 = No, 1 = Yes)",
    ylabel = "Number of passengers",
    colormap = "bwr"
)
```

**Bar chart for survived male and female**

```python
titanic_df.groupby("sex")["survived"].value_counts().plot(
    kind = "bar",
    title = "Count of survived passengers on the basis of gender",
    figsize = (10,6),
    legend = True,
    xlabel = "Passenger status (0 = dead, 1 = survived)",
    ylabel = "No of passenger",
    color = ["red","green"]
)
```

```python
titanic_df[["sex","survived"]].value_counts().plot.bar(
    color = ""
)
```

**same question using crosstab**

```python
pd.crosstab(titanic_df.survived,titanic_df.sex).plot(
    kind = "bar",
    title = "count of survived passenger",
    figsize = (10,6),
    legend = True,
    xlabel = "Passenger status (0 = dead, 1 = survived)",
    ylabel = "No of passenger",
)
```

**pie chart**

```python
group_by_embarked = titanic_df[["embarked","age"]].groupby(by = "embarked")
group_by_embarked.count().plot(
    kind = "pie",
    y = "age",
    autopct = "%1.1f%%",
    figsize = (10,6)
)
```

```python
gender_embarked = titanic_df[["sex","embarked"]].groupby(by = "embarked")
gender_embarked.count().plot(
    kind = "pie",
    y = "sex",
    figsize = (10,6),
  autopct = "%1.1f%%"
)
```

```python
survived_gender = titanic_df[["sex","survived"]].groupby(by = "sex")
survived_gender.count().plot(
    kind = "pie",
    autopct = "%1.1f%%",
    y = "survived",
    figsize = (10,6)
)
```

**pie chart of survived or not**

```python
titanic_df["survived"].value_counts().plot(
  kind = "pie",
  autopct = "%1.1f%%",
  figsize = (10,6),
  title = "Pie chart of dead and survived passengers"
)
```

**Age count by embarked**

```python
group_by_embarked = titanic_df[["embarked","age"]].groupby(by = "embarked")
group_by_embarked.count().plot(
    kind = "pie",
    autopct = "%1.1f%%",
    y = "age"
)
```

**crosstab**

```python
pd.crosstab(titanic_df["survived"],titanic_df["sex"]).plot(
    kind = "pie",
    subplots = True,
    fontsize = 10,
    figsize = (10,6),
    colormap = "Dark2"
)
```

**Line Graph**

```python
titanic_df["age"].loc[:800].plot(
    kind = "line",
    figsize = (12,8),
    color = ["red"]
)
```

```python
titanic_df["age"].loc[:1000].plot(
    kind = "line",
    figsize = (12,8),
    color = ["red"]
)
```

**line graph of age and fare**

```python
titanic_df[["age","fare"]].loc[:200].plot(
    kind = "line",
    figsize = (12,8),
    color = ["red","green"]
)
```

**scatter plot**
- helps to detect outliers

```python
titanic_df[["age","fare"]].plot(
    kind = "scatter",
    figsize = (8,6),
    title  = "Scatter plot of Age group",
    x = "age",
    y = "fare",
    color = ["green"]
)
```
