## Tuple functions

```python
# len => prints the length of tuple
print(len(tuple_2))
```

```python
# max() => prints the maximum value from the tuple
tuple_3 = (1,2,4,8,9,12,10,5)
print(max(tuple_3))
```

```python
# min() => prints the minimum value from the tuple
print(min(tuple_3))
```

```python
# sum() => prints the total sum of values from the tuple
print(sum(tuple_3))
```

```python
# any() => returns True if atleast one of the element is true
print(any(tuple_3))
```

```python
# all() => returns True if all elements are True
print(all(tuple_3))
```

```python
# sorted() => returns the sorted list from the tuple elements
print(sorted(tuple_3))
```

```python
password =""
conditions = (
    len(password) >= 8,
    '@' in password
)
if (all(conditions)):
  print("Password is valid")
else:
  print("Password is invalid")
```

```python
"Unpacking"
data = (1,3,4)
a,b,c = data
print(a)
print(b)
print(c)
```

```python
"Packing"
a = 1,2,3
print(type(a))
print(a)
```

```python
"Example of using escape variable"
digit = (1,2,3)
a, b , _ = digit
print(a)
print(b)
```

```python
" Tuple operations "
```

```python
# Indexing => used to access specific element from tuple
import random
a = (1,2,3,4,5,6,7)
num = random.randint(1,7)
print(num)
print(a[num])
```

```python
# slicing => extracting a portion of tuple using a range of indexes
print(a[1:3])
```

```python
# concatenation => combining two tuples
a = (1,2)
b = (3,4)
print(a + b)
```

```python
# repetitions => repreating the tuple elements multiple times
print(a * 2)
```

```python
# membership => checking whether an element exists in a tuple
print(3 in a)
```

```python
# comparison => comparing tuples elements by elements using comparsion operators
a = (1,2)
b = (3,4)
print(a < b)
```

```python
# here we get error because we have deleted the tuple
del a
print(a)
```

## Set

```python
"using curly brackets"
s = {'Mark','Jessa',25,75,25}
print(type(s))
print(s)
```

```python
"using set constructor"
t = set({'Mark','Jessa',25,75,25})
print(type(t))
print(t)
```

```python
"complement of set a"
a = {1,2,3,4,5,6}
b = {4,5,6}
print(a - b)
```

```python
"union"
print(a | b)
```

```python
"intersection"
print(a & b)
```

```python
"symmetric difference"
print(a ^ b)
```

```python
"subset"
print(b <= a)
```

```python
"membership"
print(4 in a)
```
