```python
""" Printing Data type of number (integer value) """
int_num = 123
print(type(int_num))
```

```python
""" Floating value """
float_num = 12.123
print(type(float_num))
```

```python
""" Complex number """
complex_num = 12 + 12j
print(type(complex_num))
```

```python
""" Converting integer into float and float into integer """
int_num = 123
float_num = float(int_num)
print(f"After converting integer into float: {float_num}.")

float_num2 = int(float_num)
print(f"After converting float into integer: {float_num2}.")
```

```python
""" Arithmetic Operations """
num1 = 22
num2 = 10

print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2}")
print(f"Modulus division: {num1 % num2}")
print(f"Exponentiation: {num1 ** num2}")
print(f"Floor division: {num1 // num2}")
```

```python
""" Comparison Operator """
a = 10
b = 20
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
""" Math functions"""
import math
```

```python
a = 99.99999
b = 10.1111
print(math.floor(a))
print(math.ceil(a))
print(math.gcd(10,30))
print(round(a,3))
print(min(a,b))
```

```python
"""Strings"""
ab = "hello world"
ac = 'hello world'
print(ab == ac)
```

```python
text1 = "It's a sunny day"
text2 = 'He said "Hello"'
print(text1)
print(text2)
```

```python
"String concatenation"
str1 = "hello"
str2 = "world"
print(str1 + " " + str2)
```

```python
"String repetition"
str3 = "hello world "
print(str3 * 3)
```

```python
"Indexing and slicing"
str4 = "Hello world!!!"
print(str4[2])

# format of slicing
""" data[start: end : step] """
print(str4[::2])
```

```python
"Reversing string using slicing"
print(str4[::-1])
```

```python
str5 = "hello python"
print(str5[1:-1])
```

```python
# spaces are also included
print(len(str5))
```
