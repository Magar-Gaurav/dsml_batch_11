# Generator

```python
def square(n):
  for i in range(n):
    return i ** 2
square(3)
```

```python
def square(n):
  for i in range(n):
    yield i ** 2
a = square(3)
```

```python
print(next(a))
```

```python
def count_up(n):
  while n <= 10:
    yield n
    n += 1
a = count_up(1)
```

```python
print(next(a))
```

# **Exception Handling**

**Example**

```python
# zero division error
def divide_numbers(num1,num2):
  try:
    result = num1 / num2
    print(f"Result of {num1} / {num2} is {result}.")
  except ZeroDivisionError:
    print("Error: Division by Zero not allowed.")
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
divide_numbers(num1,num2)
```

```python
# type error
a = input("Enter num1: ")
b = int(input("Enter num2: "))
try:
  sum = a + b
  print(f"Sum of {a} and {b} is {sum}")
except TypeError:
  print(f"Error: Two different type")
```

```python
# value error
user_inp = input("Enter your input value: ")
try:
  num = int("abc")
  print(num)
except ValueError:
  print("Error: Value Error")
```

```python
# Index error
list2 = [1,2,3,4]
try:
  print(f"Value: {list2[4]}")
except IndexError:
  print("Invalid index: Index out of range")
```

```python
# Key error
dict1 = {
    "name" : "Saurav"
}
try:
  print(dict1["age"])
except KeyError:
  print("Error: Age key not found")
```

```python
# file not found
try:
  with open("file.txt","r") as f:
    data = f.read()
    print(data)
except FileNotFoundError:
  print("File not found")
```

```python
# file upload
try:
  with open("new_file.txt","w") as f:
    f.write("Hello user this content was written and error is handled using exception handling.")
except FileNotFoundError:
  print("Error: File not found")
finally:
  f.close()
```

# **Custom Error**

```python
def check_age(age):
  if age < 0:
    raise ValueError("Age cannot be negative")
  print("Age is ",age)
```

```python
try:
  age = int(input("Enter your age: "))
  check_age(age)
except ValueError as e:
  print(e)
```

# Custom Exception example

```python
class InvalidAgeError(Exception):
  """Custom exception for invalid age values"""
  pass
def set_age_custom(age):
  if age > 120:
    raise InvalidAgeError("Age canot be over 120")
```

```python
try:
  age = int(input("Enter your age: "))
except Exception as e:
  set_age_custom(age)
```

# Class Task

Develop a mini ATM program that uses custom exceptions, try-except and finally to handle deposits, withdrawls and balance checks safely.

```python
def check_balance(balance,amt):
  if balance <= 0:
    raise ValueError("Balance is less than or equal to 0 cannot withdraw.")
  elif balance < amt:
    raise ValueError("Balance amount is less than the amount to withdraw.")
def check_amount(amount):
  if amount < 1:
    raise ValueError("Amount less than 1 cannot be deposited.")
```

```python
print("Welcome to Mini ATM Program")
balance = 0
while True:
  print("=="*50)
  print("1. Deposit")
  print("2. Withdraw")
  print("3. Balance Check")
  print("4. Exit")
  user_choice = int(input("Enter your choice: "))
  if user_choice == 1:
    try:
      amt = int(input("Enter the amount to deposit: "))
      check_amount(amt)
      balance += amt
    except ValueError as e:
      print(e)
    finally:
      print(f"Your current balance is Rs.{balance}.")
  elif user_choice == 2:
    amt = int(input("Enter the amount to withdraw: "))
    try:
      check_balance(balance,amt)
      balance -= amt
    except ValueError as e:
      print(e)
    finally:
      print(f"Your current balance is Rs. {balance}.")
  elif user_choice == 3:
    print(f"Your current balance is Rs.{balance}.")
  elif user_choice == 4:
    print("Exitting....")
    break
  else:
    print("Invalid choice")
    continue
```
