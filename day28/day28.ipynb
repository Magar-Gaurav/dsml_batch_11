# **Practice Question**

- **Analyze survival by :**
  - Gender
  - Passenger class
  - Age group
  - Fare
- **Based on you findings, describe the characteristics of a passenger with the highest chance of survival.**

```python
titanic_df.columns
```

**Survival by Gender**

```python
titanic_df[["Survived","Sex"]].value_counts(normalize=True,ascending=True)
```

**So, from the above proportion we can see that the female are likely more to survive than men.**

**Survival by Passenger Class**

```python
titanic_df[["Survived","Pclass"]].value_counts(normalize = True,ascending=True)
```

**So, from the above proportion we can see 1st class people are more survived than other classes**

**Survival by Age Group**

```python
titanic_df["AgeGroup"] = pd.cut(
    titanic_df["Age"],
    bins=[0, 12, 18, 40, 60, 80],
    labels=["Child","Teen","Adult","Middle-aged","Senior"]
)

titanic_df[["Survived","AgeGroup"]].value_counts(normalize=True, ascending=True)
```

**So, from the above proportion or description we can say that Adult group are more likely to survive and high rate of no survival.**

```python
titanic_df["FareGroup"] = pd.cut(
    titanic_df["Fare"],
    bins=[0, 7.91, 14.45, 31, 513],
    labels=["Low", "Medium", "High", "Very High"]
)

titanic_df[["Survived","FareGroup"]].value_counts(normalize=True, ascending=True)
```

**From above description we can asy that high paying fare group people have high chances of surviving**

**Task2: Embarkation Port Analysis**

**The company wants to know which port contributed the most passengers.**

**Tasks:**
- Count passengers from each embarkation port (Embarked)
- Calculate the survival rate for each port
- Identify the port with the highest survival rate

**Counting passengers from each embarked port**

```python
titanic_df.value_counts("Embarked")
```

**Survival rate for each port**

```python
titanic_df.groupby("Embarked")["Survived"].mean()
```

**Port with highest survival rate**

```python
titanic_df.groupby("Embarked")["Survived"].mean().max()
```

**People from C port has highest survival rate which is 0.5536%**
