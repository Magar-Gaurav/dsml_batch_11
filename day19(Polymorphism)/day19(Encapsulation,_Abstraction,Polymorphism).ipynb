**Create a bank account class with :**

**Create a SavingAccount class that adds 5% interest to the balance using methods**

```python
class BankAccount:
  def __init__(self,a_name,a_type,u_pin,a_balance):
    self.a_name = a_name
    self._a_type = a_type
    self.__u_pin = u_pin
    self.__a_balance = a_balance
  def get_balance(self):
    return self.__a_balance
  def set_balance(self,amt):
    self.__a_balance = amt
  def acc_dets(self):
    print(f"Account holder name: {self.a_name}\nAccount Type: {self._a_type}\nBalance: Rs.{self.__a_balance}.")
```

```python
class SavingAccount(BankAccount):
  def __init__(self,a_name,a_type,u_pin,a_balance,rate):
    super().__init__(a_name,a_type,u_pin,a_balance)
    self.rate = rate
  def interest_balance(self):
    self.acc_dets()
    amt = ((self.rate /100) *   self.get_balance())
    print(f"After {self.rate}% interest on your balance is Rs.{self.get_balance() + amt}.")
```

```python
cust1 = SavingAccount("Saurav","saving","5001",50000,5)
cust1.interest_balance()
```

**Create an abstract BankAccount class with abstract methods**
- deposit()
- withdraw()

**Implement them in a SavingAccount class**

```python
from abc import abstractmethod
class BankAccount:
  def __init__(self,balance):
    self.__balance = balance
  @abstractmethod
  def deposit(self):
    pass
  @abstractmethod
  def withdraw(self):
    pass
  def get_balance(self):
    return self.__balance
  def set_balance(self,amt):
    self.__balance = amt
  def display_balance(self):
    print(f"Balance: {self.__balance}")
```

```python
class SavingAccount(BankAccount):
  def __init__(self,balance,d_amt,w_amt):
    super().__init__(balance)
    self.d_amt = d_amt
    self.w_amt = w_amt
  def deposit(self):
    self.display_balance()
    if self.d_amt >= 0:
      self.set_balance(self.d_amt)
      print(f"Successfully deposited Rs.{self.d_amt}.")
    else:
      print("Cannot deposit amout less than or equal to 0.")
  def withdraw(self):
    main_balance = self.get_balance()
    if main_balance >= self.w_amt:
      print(f"Successfully withdrawn Rs.{self.w_amt}.")
    else:
      print("Insufficient Balance.")
    self.display_balance()
```

```python
balance = int(input("Enter your initial balance: "))
deposit_amt = int(input("Enter the amount you want to deposit: "))
withdraw_amt = int(input("Enter the amount to withdraw: "))
```

```python
user1 = SavingAccount(balance,deposit_amt,withdraw_amt)
user1.deposit()
user1.withdraw()
```

#     **Polymorphism**

```python
class PaymentGateway:
  def process_payment(self,amount):
    pass

class CreditCard(PaymentGateway):
  def process_payment(self,amount):
    return f"Processing ${amount} via Credit Card authentication."

class PayPal(PaymentGateway):
  def process_payment(self,amount):
    return f"Processing ${amount} via PayPal secure token."

# unified interface function
def execute_checkout(payment_method,amount):
  print(payment_method.process_payment(amount))

# usage
card = CreditCard()
wallet = PayPal()

execute_checkout(card,5000)
execute_checkout(wallet,1000)
```

**Opeartor Overloading**

```python
class Student:
  def __init__(self,m1,m2):
    self.m1 = m1
    self.m2 = m2
  def __add__(self,other):
    m1 = self.m1 + other.m1
    m2 = self.m2 + other.m2
    s3 = Student(m1,m2)
    return s3
```

```python
std1 = Student(10,20)
std2 = Student(30,19)
std3 = std1 + std2
```

```python
class Comparison:
  def __init__(self,num1,num2):
    self.num1 = num1
    self.num2 = num2
  def __gt__(self,other):
    total1 = self.num1 + self.num2
    total2 = other.num1 + other.num2

    if total1 > total2:
      return True
    else:
      return False
```

```python
comparison = Comparison(15,10)
comparison2 = Comparison(15,15)
comparison1 = comparison > comparison2
print(comparison1)
```
