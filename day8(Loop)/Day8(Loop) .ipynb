```python
"sum of 10 number"
sum = 0
i = 1
while i <= 10:
  sum += i
  i += 1
print("sum of 10 number is ", sum)
```

```python
" Example of infinite loop "
while True:
  print("Hello world")
```

```python
"Example of break"
i = 1
while i <= 10:
  print(i)
  if i == 5:
    break
  i += 1
```

```python
"Example of continue"
i = 1
while i <= 10:
    if i == 5:
        i += 1
        continue
    print(i)
    i += 1
```

```python
" Example of pass"
i = 1
while i <= 10:
    if i == 5:
        pass
    print(i)
    i += 1
```

```python
"for loop"
for i in range(10):
  print(i)
```

```python
"from 1 to 20"
for i in range(1,21):
  print(i)
```

```python
"Even number"
for i in range(2,21,2):
  print(i)
```

```python
# C.w: take user input continously (7 as guess number) if they get the correct one print correct
guess_number = 7
while True:
  user_choice = int(input("Enter your guess: "))
  if user_choice == guess_number:
    print("Correct")
    break
  else:
    print("Incorrect")
```

```python
"Ask user for 5 numbers"
positive_sum = 0
for i in range(5):
  user_input = int(input("Enter a number: "))
  if user_input > 0:
    positive_sum += user_input
print("Sum of positive numbers is ", positive_sum)
```

```python
"find positive and negative number sum with their own list"
positive_sum = 0
negative_sum = 0
negative_list = []
positive_list = []
for i in range(5):
  user_input = int(input("Enter a number: "))
  if user_input > 0:
    positive_list.append(user_input)
    positive_sum += user_input
  else:
    negative_list.append(user_input)
    negative_sum += user_input
print(f"Negative List: {negative_list}")
print(f"Total negative sum: {negative_sum}")
print(f"Positive List: {positive_list}")
print(f"Total positive sum: {positive_sum}")
```

```python
"Create a simple ATM Menu: 1. Check balance 2. Deposit 3. Exit"
balance = 1000
print("\t\tWelcome to the Simple ATM!!!")
while True:
  user_choice = int(input("Enter your choice (1 for Checking Balance, 2 for deposit, 3 for exit): "))
  if user_choice == 1:
    print(f"Your current balance is Rs.{balance}.")
  elif user_choice == 2:
    deposit_amt = float(input("Enter the amount to deposit: "))
    if deposit_amt > 0:
      balance += deposit_amt
      print(f"Successfully deposited Rs.{deposit_amt} into your account.\nYour currrent balance is Rs.{balance}.")
    else:
      print("Cannot deposit amount less than 1.")
  elif user_choice == 3:
    print("Exiting....")
    break
  else:
    print("Please enter a valid choice!!!")
```
