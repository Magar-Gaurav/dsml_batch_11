# **Pandas Basics**

**importing pandas**

```python
import pandas as pd
```

**Example 1: Converting a list into a series**

```python
a = ['g','h','i','j']
res = pd.Series(a)
print(res)
print(type(res))
```

**Syntax of series**

```python
a = ['1','2','3','4','5']
res = pd.Series(a,index=['a','b','c','d','e'],dtype="int8",name = "list of numbers",copy=False)
print(res)
```

**Example 2: Converting dictionary to Series**

```python
a = {
    "name" : "Saurav",
    "age" : 23,
    "faculty" : "BCA"
}
df = pd.Series(a)
print(df)
print(type(df))
```

**DataFrame**

**Example 1: Creating DataFrame using List**

```python
data = ["Python","Pandas","NumPy"]
df = pd.DataFrame(data,columns = ["Tech"],index = ['a','b','c'])
print(df)
```

**Example 2: Creating DataFrame using Dictionary**

```python
a = {
    "Student Name" : ["Saurav","Gaurav","Ram","Hari"],
    "Age" : [23,32,12,21],
    "Faculty" : ["BCA","BCA","CsIT","Btech"]
}
df = pd.DataFrame(a,index=["a","b","c","d"],dtype="str",columns=["Student Name","Age","Faculty"],copy=False)
print(df)
```

**Fetching the Student Name column from the dataframe**

```python
df[["Student Name","Age"]]
```
