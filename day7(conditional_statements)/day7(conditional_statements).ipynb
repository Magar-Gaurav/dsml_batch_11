```python
#Example of if-else block
"Example of checking if age is above 18"
age = 13
if age > 18:
  print("Age is greater than 18")
else:
  print("Age is less than 18")
```

```python
# Example of if-elif-else block
"Asking for user to input the age"
age = int(input("Enter your age: "))
if age > 18 and age < 100:
  print("You can vote!!!")
elif age > 0 and age <= 18:
  print("You cannot vote!!!")
else:
  print("Please enter a valid age!!!")
```

```python
# ask for age and check citizenship
age = int(input("Enter your age: "))
have_citizenship = input("Do you have citizenship card? (y for yes and n for no): ")
if age >= 18 and have_citizenship == "y":
  print("You are eligible to vote")
elif age >= 18 and have_citizenship == "n":
  print("You are eligible to vote but create your citizenship card first")
else:
  print("You are not eligible to vote")
```

```python
# Class Task
"Write a program to take marks as input from the user and display the grade using if and elif statements according to the following \
 90 to 100 => A+ \
 80 to 90 => A \
 70 -80 => B+ \
 60- 70 => B\
 50 - 60 => C+\
 40 - 50 => C\
 30 - 40 => D\
 < 30 => Fail\
"
marks_obtained = int(input("Enter marks obtained: "))
if 100 > marks_obtained >= 90:
  print("A+")
elif 90 > marks_obtained >= 80:
  print("A")
elif 80 > marks_obtained >= 70:
  print("B+")
elif 70 > marks_obtained >= 60:
  print("B")
elif 60 > marks_obtained >= 50:
  print("C+")
elif 50 > marks_obtained >= 40:
  print("C")
elif 40 > marks_obtained >= 30:
  print("D")
else:
  print("NG")
```

```python
"ask for age :  (age > 40 and membership condition check)\
age = 40 + membership => gets 60% discount\
age = 40+ no membership => gets 40% discount\
else: no discount"
user_age = int(input("Enter your age: "))
membership = input("Enter your membership (yes / no):")
if user_age >= 40 and membership == "yes":
  print("You get 60% discount")
elif user_age >= 40 and membership == "no":
  print("You get 40% discount")
else:
  print("You get no discount")
```

```python
"create a list and ask for user input and if it is even add to the list else delete one element from the list"
lst = [3,4,6,7,9]
user_input = int(input("Enter the number: "))
lst.append(user_input) if user_input % 2 == 0 else lst.pop()
print(lst)
```
