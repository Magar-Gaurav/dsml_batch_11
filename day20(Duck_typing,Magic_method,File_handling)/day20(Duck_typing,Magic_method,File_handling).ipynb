**Create classes Esewa, Khalti and BankTransfer each having a pay(self.amount) method. Write a payment processor function that works with any payment method object.**

```python
class Esewa:
  def __init__(self,amount):
    self.amount = amount
  def pay(self):
    return f"Rs.{self.amount} successfully paid through e-sewa."
class Khalti:
  def __init__(self,amount):
    self.amount = amount
  def pay(self):
    return f"Rs.{self.amount} successfully paid through khalti."
class BankTransfer:
  def __init__(self,amount):
    self.amount = amount
  def pay(self):
    return f"Rs.{self.amount} successfully paid through Bank Transfer."

def payment(payment_method):
  print(payment_method.pay())
```

```python
amt = int(input("Enter the amount to pay: "))
```

```python
esewa = Esewa(amt)
khalti = Khalti(amt)
bankTransfer = BankTransfer(amt)
```

```python
payment(esewa)
```

```python
payment(khalti)
```

```python
payment(bankTransfer)
```

```python
choice = input("Enter your choice (e for e-sewa, k for khalti, b for bankTransfer): ").strip().lower()
if choice == "e":
  payment(esewa)
elif choice == "k":
  payment(khalti)
elif choice == "b":
  payment(bankTransfer)
else:
  print("Please enter a valid choice!!!")
```

# **magic methods**

- special method with names covered with "__" (double underscore)

```python
class BankCustomer:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # called by print
  def __str__(self):
    return f"{self.name} is {self.age} years old."

  # called by repr()
  def __repr__(self):
    return f"BankCustomer : ('{self.name}', {self.age})"

  # called by len()
  def __len__(self):
    return self.age

  # makes object callable like a function
  def __call__(self,gender = "Male"):
    return f"{self.name} is a {gender}."
```

```python
cust1 = BankCustomer("Saurav",23)
```

```python
print(cust1.__str__())
```

```python
print(cust1.__repr__())
```

```python
print(cust1.__len__())
```

```python
print(cust1.__call__())
```

```python
# create object
customer = BankCustomer("Ram",23)

# __str__
print(customer)

# __repr__
print(repr(customer))

# __len__
print(len(customer))

# __call__
print(customer())
```

**Create a BankAccount class and implement __str__() and __repr__() methods to display account details in user-friendly and developer-friendly formats.**

```python
class BankAccount:
  # constructor
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # __str__()
  def __str__(self):
    return f"{self.name} is of {self.age} years old."

  # __repr__()
  def __repr__(self):
    return f"Account Holder Name: {self.name}\nAccount Holder Age: {self.age}."
```

```python
acc_name = input("Enter the name of account holder: ")
acc_age = int(input("Enter the age of account holder: "))
```

```python
cust1 = BankAccount(acc_name,acc_age)
print(cust1)
```

```python
print(repr(cust1))
```
