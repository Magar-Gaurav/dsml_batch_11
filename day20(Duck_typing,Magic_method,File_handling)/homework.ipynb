**Find words from web and then split them sentence by sentence and store them in excel file (in the format: serial number | sentence)**

```python
import csv

text = "Defending champions Argentina began their World Cup title defense in emphatic fashion with a convincing victory over Algeria, inspired by a historic hat-trick from Lionel Messi. Playing his 200th international match, Messi scored the first World Cup hat-trick of his career and became the joint-highest goalscorer in World Cup finals history with 16 goals, equaling the record held by Miroslav Klose. The eight-time Ballon d'Or winner opened the scoring in the 17th minute with a brilliant curling effort after receiving a pass from teammate Rodrigo De Paul. Argentina faced strong resistance from Algeria, but Messi doubled the lead shortly after halftime by capitalizing on a mistake from goalkeeper Luca Zidane, son of French legend Zinedine Zidane. He completed his hat-trick in the 76th minute with a powerful strike from the edge of the box, sealing a memorable win. The victory marked Argentina's first-ever opening-match win at a World Cup as defending champions and provided a strong start to their Group J campaign, with upcoming matches against Austria and Jordan."

sentences = text.split(".")
try:
    with open("homework.csv","w",newline = "") as f:
        csv_writer = csv.writer(f)
        csv_writer.writerow(["Serial Number", "Sentence"])

        for i,sentence in enumerate(sentences,start = 1):
            if sentence.strip():
                csv_writer.writerow([i, sentence])
    print("Sentences saved to homework.csv successfully.")
except Exception as e:
    print("Error occured: ",e)
```

**converting the csv file into excel**

```python
import pandas as pd
try:
    df = pd.read_csv("homework.csv")
    df.to_excel("homework.xlsx",index=False)
    print("Successfully converted csv file into the excel format.")
except Exception as e:
    print(f"Error {e} has occured!!!")
```
