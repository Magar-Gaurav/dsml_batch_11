# Recursion

```python
# finding factorization of a number using recursion
def factorial(n):
  if n == 0 | n == 1:
    return 1
  return n * factorial(n - 1)
print(f"Factorial of 5 is {factorial(5)}.")
```

```python
# finding fibonacci series using recursion
def fibonacci(n):
  if n <= 1:
    return n
  return (fibonacci(n - 1) + fibonacci(n-2))
n = int(input("Enter the number of series you need: "))
print(f"Fibonacci series till {n}th terms: ")
for i in range(n):
  print(fibonacci(i),end= " ")
```

# Nested function

```python
# how to pass value to inner and outer function
def outer(x):
  def  inner(y):
    return x + y
  return inner
add_ = outer(5)
res = add_(3)
print(res)
```

# Passing function as an argument

```python
def add(x,y):
  return x + y
def calculate(func,x,y):
  return func(x,y)
result = calculate(add,5,4)
print(result)
```

```python
# square
def square_num(x):
  return x * x
def calculate(func,x):
  return func(x)
num = int(input("Enter the number whose square you need: "))
result = calculate(square_num,num)
print(f"Square of {num} is {result}.")
```

# Passing function as a value

```python
def greeting(name):
  def hello():
    return "Hello " + name + "!"
  return hello
greet = greeting("Saurav")
print(greet())
```

# Decorator

```python
def myWrapper(func):
  def myFunc():
    print("Hello World! {From inside wrapper}")
    func()
  return myFunc
@myWrapper
def myFunc():
  print("Hellow World! {From outer wrapper}")
myFunc()
```

```python
# decorator with arguments
def mainWrapper(func):
  def printName(*args,**kwargs):
    print(f"Calling function : {func.__name__}")
    return func(*args,*kwargs)
  return printName

@mainWrapper
def printName(name,age):
  print(f"Hello {name}, Age: {age}.")
printName("Ram", 23)
```

```python
# Before function execution
# Hello {name}
# After function execution
def mainWrapper(func):
  def innerWrapper(*args):
    print("Before function execution")
    return func(*args)
  return innerWrapper

@mainWrapper
def innerWrapper(name):
  print(f"Hello {name}!")
  print("After function execution")

innerWrapper("Ram")
```
