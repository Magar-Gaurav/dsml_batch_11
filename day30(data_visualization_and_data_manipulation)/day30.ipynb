# **Central Tendency**

**importing dependencies**

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = np.random.standard_t(500,size = 1000)
x[:15]
```

```python
_mean = np.mean(x)
print(f"Mean of x: {_mean}")
```

```python
plt.hist(x,bins = 10,color = "blue")
plt.axvline(x = _mean,color = "yellow")
```

```python
plt.axvline(x = _mean,color = "yellow")
```

**median and mode task**

**mode**

```python
mode_result = stats.mode(x)
mode = mode_result.mode
print(mode)
plt.hist(x, bins=10, color="blue")
plt.axvline(x=mode, color="green")
plt.show()
```

```python
sns.histplot(x,bins = 10,color = "black",kde = True,edgecolor=None)
plt.axvline(x = mode,color = "green")
```

**median**

```python
median = np.median(x)
print("Median of x:",median)
```

```python
plt.hist(x,bins = 10,color = "blue",edgecolor = "black")
plt.axvline(median,color = "red")
plt.ylabel("Count")
plt.xlabel("range")
plt.show()
```
