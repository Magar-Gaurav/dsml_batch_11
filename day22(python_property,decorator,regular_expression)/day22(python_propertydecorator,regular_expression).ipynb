# **Python Property Decorator**

**@property**

```python
class Student:
    def __init__(self,name):
        self._name = name

    @property
    def name(self):
        return self._name
```

```python
s = Student("Ram")
s.name
```

**getter**

```python
class Student:
    def __init__(self,name):
        self._name = name
    @property
    def name(self):
        print("Getter is called")
        return self._name
```

```python
std = Student("Gaurav")
std.name
```

**setter**

```python
class Student:
    def __init__(self,age):
        self._age = age
    @property
    def age(self):
        print("Getter is called")
        return self._age
    @age.setter
    def age(self,value):
        print("Setter called")
        if value < 0:
            raise ValueError("Age cannot be less than or equal to 0.")
        self._age = value
```

```python
s = Student(23)
s.age
```

```python
s.age = 40
```

```python
class Student:
    def __init__(self,name):
        self._name = name
    @property
    def name(self):
        print("Getter called")
        return self._name
    @name.setter
    def name(self,value):
        print("Setter called")
        if len(value) < 3:
            raise ValueError("Name cannot be less than 3.")
        self._name = value
```

```python
std1 = Student("Ram")
std1.name
```

```python
std1.name = "Shyam"
std1.name
```

**Deleter**


```python
class Student:
    def __init__(self,age):
        self._age = age
    @property
    def age(self):
        return self._age
    @age.deleter
    def age(self):
        print("Deleter calling....")
        del self._age
```

```python
s = Student(20)
del s.age
```

**Class task: Using getter, setter and deleter**

```python
class Student:
  def __init__(self,name,age):
    self._name = name
    self._age = age
  @property
  def name_age(self):
    print("Getter called")
    return self._name,self._age
  @name_age.setter
  def name_age(self,values):
    print("Setter called")
    self._name,self._age = values
  @name_age.deleter
  def name_age(self):
    print("Deleter calling......")
    del self._name
    print("Successfully deleted name")
```

```python
student = Student("Ram",23)
student.name_age
```

```python
student.name_age = ("Ram",20)
student.name_age
```

```python
del student.name_age
```

**Regex**

**search**

```python
import re

text = '''Python is awesome. I love Python because Python is easy to learn. You can use Python for AI, Web Development, and Datascience.'''
match_result = re.search("Python",text)
match_result
```

```python
if match_result:
  print("Match found for:",match_result.group())
else:
  print("No match found")
```

**find_all**

```python
import re

text = '''Python is awesome. I love Python because Python is easy to learn. You can use Python for AI, Web Development, and Datascience.'''
match_result = re.findall("Python",text)
print(match_result)
```

**Web Scrapping**

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Choose a webpage
url = "https://en.wikipedia.org/wiki/Web_scraping"   # Example site

# Step 2: Fetch the webpage
response = requests.get(url)
html_content = response.text

# Step 3: Parse the HTML
soup = BeautifulSoup(html_content, "html.parser")

# Step 4: Extract text (e.g., all <p> tags)
paragraphs = soup.find_all("p")

# Step 5: Print the text
for i, p in enumerate(paragraphs, start=1):
    print(f"{i}. {p.get_text(strip=True)}")
```
